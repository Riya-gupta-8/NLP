Some Text Preprocessing.I am not covering all relevant preprocessing as the purpose of this notebook is on the Modelling approach

```python
# Text Processing 
#1. Lower casing
#2. Stop words
#3. Remove the tags  
```

```python
# remove the tags 

import re
def remove_tags(text):
    cleaned_text = re.sub(re.compile("<.*?>") , '' , text)
    return cleaned_text
```

```python
df['review'] = df['review'].apply(remove_tags)
```

```python
df.head()
```

```python
# convert to the lower case the all text 
df['review'] = df['review'].apply(lambda x:x.lower())
```

Let's use the Word2Vec text embeddings and build the model 

```python
%pip install gensim
```

```python
import gensim
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
df.head()
```

```python
#The error you're encountering (LookupError) typically occurs when NLTK is unable to find the data it needs for tokenization. Specifically, in this case, it seems NLTK is unable to find the tokenizer data for the language you're using.
#To resolve this, you can download the necessary NLTK data. You'll need to download the Punkt tokenizer models for the English language or any other language you might be using.

nltk.download('punkt')
```

```python
story = []
for doc in df['review']:
    raw_sent = sent_tokenize(doc)
    for sent in raw_sent:
        story.append(simple_preprocess(sent))
```

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2
)
```

```python
model.build_vocab(story)
```

```python
model.train(story, total_examples=model.corpus_count, epochs= model.epochs)
```

```python
len(model.wv.index_to_key)
```

```python
def document_vector(doc):
    doc = [word for word in doc.split() if word in model.wv.index_to_key]
    return np.mean(model.wv[doc] , axis = 0)
```

```python
document_vector(df['review'].values[0])
```

```python
from tqdm import tqdm 
```

```python
X = []
for doc in tqdm(df['review'].values):
    X.append(document_vector(doc))
```

```python
X = np.array(X)
X.shape
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

y = le.fit_transform(df['sentiment'])
```

```python
y
```

Modelling using random forest classifier 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred = rfc.predict(X_test)
accuracy_score(y_test,y_pred)
```
